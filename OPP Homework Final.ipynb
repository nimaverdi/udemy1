```python
#1) using tuple unpacking

class Line2():
    
    def __init__(self,coor1,coor2):
        self.coor1=coor1
        self.coor2=coor2
    
    def distance_2(self): #using tuple unpacking, it's cleaner!
        (x1,y1) = self.coor1
        (x2,y2) = self.coor2
        diff1=abs(y2-y1)
        diff2=abs(x2-x1)
        
        length=math.sqrt(diff1**2 + diff2**2)
        return length
    
    def slope(self): #apply average rate of change: change in y per change in x
        (x1,y1) = self.coor1
        (x2,y2) = self.coor2
        delta_y=y2-y1
        delta_x=x2-x1
        
        roc=delta_y/delta_x
        return roc
    
        
    
test = Line2((1,1),(3,3))
test.distance_2()
test.slope()

#works
```

```python
#2)

import math #so we can use math.pi

class Cylinder():
    
    def __init__(self,height=1,radius=1): #default inputs of 1
        self.height = height
        self.radius = radius
        
    def volume(self):
        vol = math.pi * self.radius**2 * self.height
        return vol
    
    def surface_area(self):
        surf_area = 2 * math.pi * self.radius * self.height + 2 * math.pi * self.radius**2
        return surf_area

test_cylinder = Cylinder(2,3)
test_cylinder.volume()
test_cylinder.surface_area() #works
```
