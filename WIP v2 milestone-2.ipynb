```python
#1 print board

def display_board(board):
    
    print( board[7] + '|' + board[8] + '|' + board[9])
    print( board[4] + '|' + board[5] + '|' + board[6])
    print( board[1] + '|' + board[2] + '|' + board[3])
    
    
test_board=['#','X','O','X','O','X','O','X','O','X']
display_board(test_board)

#works
```

```python
#4 check to see if mark won
board = ['#','X','O','X','O','X','O','O','O','X']

def win_check(board,mark):
    if board[1:4]==list(3*mark) or board[4:7]==list(3*mark) or board[7:9]==list(3*mark) or board[1:10:4]==list(3*mark) or board[3:8:2]==list(3*mark) or board[1:10:3]==list(3*mark) or board[2:10:3]==list(3*mark) or board[3:10:3]==list(3*mark):
        print(True)
    else:
        print(False)
        
win_check(test_board,'X')

#works
```

```python
#5 function randomly deciding which player goes first

import random

def choose_first():
    player=random.randint(1,2)
    if player==1:
        print('Player 1, you go first!')
    else:
        print('Player 2, you go first!')
    
choose_first() #works
```

```python
#6 check if space on board avail

def space_check(board,position):
    pos=int(position)
    if board[pos]=='':
        return True
    else:
        return False
    
# returns True if space avail and False otherwise
    
space_check(test_board,2) #works
```

```python
#7 check if board is full

def full_board_check(board):
    short_board=board[1::]
    x_count=short_board.count('X')
    o_count=short_board.count('O')
    board_moves=x_count+o_count
    if board_moves==9:
        return True
    else:
        return False
    
    
full_board_check(test_board) #works
```
