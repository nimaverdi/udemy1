```python
#class is a blueprint/rule to create custom object
# instance of an object is a specific object created from particular class (blueprint)

class Sample():
    pass

my_sample=Sample() #instance of class
type(my_sample)
```

```python
# define a class of objects "Dog" which has attributes

class Dog():
    def __init__(self,breed,name,spots): #init is a class constructer, creates instance of class. it is a special method which is automatically called when an object of that Class is created.
        #here we relate attributes (.breed) to the arguments above
        self.breed = breed #input for breed is being associated with .breed attribute
        self.name = name
        self.spots = spots
        
#now we can create an instance of the object from the class

my_dog = Dog(breed='Lab',name='aziz',spots=True)

#now if we check the attributes of this object, we will see the arguments we associated

my_dog.breed #returns 'Lab'
#generally, check attributes/methods from my_dog. and tab+shift
```

```python
# OPP Part 2: Class Object Attributes

#so far we have defined attributes at the object level
#if there are attributes that are the same across all instances of object, we can define class-level attributes (class object attribute)
#for example, all dogs are mammals

class Dog2():
    
    species = 'mammal' #define class object attribute here
    
    def __init__(self,breed,name,spots): 
        
        self.breed = breed
        self.name = name
        self.spots = spots

my_dog2=Dog2(breed='husky',name='sammy',spots=False)
my_dog2.species
```

```python
#METHODS

#functions that perform operations on object
#defined inside class

class Dog3():
    
    species = 'mammal' #define class object attribute here
    
    def __init__(self,breed,name,spots): 
        
        self.breed = breed
        self.name = name
        self.spots = spots
        
    def bark(self):
        print('Woof!')

my_dog3 = Dog3('bernese','chow',False)
my_dog3.bark() #this will execute the method "bark", just like "append" etc
#note, we need brackets to execute method

#see last few minutes of Part 2 lecture for a good example

#SUMMARY
#OPP allows users to create custom data types (objects) with their own attributes (characteristics) and methods (pre-built functions)
#objects are defined using Class key
```

```python
#Part 3: INHERITANCE & POLYMORPHISM (optional)

#Inheritance

#new classes can assume methods and attributes of already existing classes
#benefit: re-use code

#----------------------------example------------------

#define a base class

class Animal():
    
    def __init__(self): #__init__ method automatically called when instance of the class created, this is why it Animal() returns 'ANIMAL CREATED'
        print('ANIMAL CREATED')
    
    def who_am_i(self):
        print('I am an animal')
        
    def eat(self):
        print('I am eating')

        
#we can now define new class which inherits attributes/methods of Animal() class

class Dog(Animal): #enter base class name in brackets of new class
    
    def __init__(self):
        Animal.__init__(self) #creating instance of Animal class from instance of Dog class
                                #it's executing a function
        
    #can also overrwrite existing method:
    def eat(self):
        print('I am a dog and i am eating')
        
    #can add new methods as usual
        
my_dog_new=Dog()

#now my_dog uses same attributes/methods as Animal:

my_dog_new.who_am_i()
my_dog_new.eat()
```

```python
#Part 4 - SPECIAL METHODS (magic/dunder methods)

# how to use built in Python methods with my own user defined objects

class Book():
    
    def __init__(self,title,author,pages):
        self.title=title
        self.author=author
        self.pages=pages
        
    def __str__(self): #creates a string representation of object so can print() object
        return self.title
    
    
    def __len__(self): #length of user defined objects
        return self.pages
    
b=Book('Python rocks','Me',200)

```
