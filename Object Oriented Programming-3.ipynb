```python
#OPP

#allows users to create custom objects (data types) with their own methods (operations) and attributes (characteristics)
#used to create repeatable and organized code (functions aren't enough at large scale code)

#general syntax:

# class NameOfClass():
    #def__init__(self,param1,param2):
        #self.param1=param1
        #self.param2=param2
    
#define object using class keyword and give it name  (follows capitalized wording)
```

```python
#class is a blueprint/rule to create custom object
# instance of an object is a specific object created from particular class (blueprint)

class Sample():
    pass

my_sample=Sample() #instance of class
type(my_sample)
```

```python
# define a class of objects "Dog" which has attributes

class Dog():
    def __init__(self,breed,name,spots): #init is a class constructer
        #here we relate attributes (.breed) to the arguments above
        self.breed = breed #input for breed is being associated with .breed attribute
        self.name = name
        self.spots = spots
        
#now we can create an instance of the object from the class

my_dog = Dog(breed='Lab',name='aziz',spots=True)

#now if we check the attributes of this object, we will see the arguments we associated

my_dog.breed #returns 'Lab'
#generally, check attributes/methods from my_dog. and tab+shift
```

```python
# OPP Part 2: Class Object Attributes

#so far we have defined attributes at the object level
#if there are attributes that are the same across all instances of object, we can define class-level attributes (class object attribute)
#for example, all dogs are mammals

class Dog2():
    
    species = 'mammal' #define class object attribute here
    
    def __init__(self,breed,name,spots): 
        
        self.breed = breed
        self.name = name
        self.spots = spots

my_dog2=Dog2(breed='husky',name='sammy',spots=False)
my_dog2.species
```

```python
#METHODS

#functions that perform operations on object
#defined inside class

class Dog3():
    
    species = 'mammal' #define class object attribute here
    
    def __init__(self,breed,name,spots): 
        
        self.breed = breed
        self.name = name
        self.spots = spots
        
    def bark(self):
        print('Woof!')

my_dog3 = Dog3('bernese','chow',False)
my_dog3.bark() #this will execute the method "bark", just like "append" etc
#note, we need brackets to execute method

#see last few minutes of Part 2 lecture for a good example

#SUMMARY
#OPP allows users to create custom data types (objects) with their own attributes (characteristics) and methods (pre-built functions)
#objects are defined using Class key
```
