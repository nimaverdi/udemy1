```python
class Account():
    
    def __init__(self,owner,balance):
        self.owner = owner
        self.balance = balance
        
    def deposit(self,deposit_amt): #methods can accept arguments like regular functions. They need "self" to specify that method performed on instance of class
        new_balance_d= self.balance + deposit_amt
        return new_balance_d
    
    def withdraw(self,withdraw_amt):
        new_balance_w = self.balance - withdraw_amt
        if new_balance_w < 0:
            print('Withdrawal exceeds balance')
        else:
            return new_balance_w
    
test = Account('Nima',100)

test.deposit(20) #we need to specify the deposit amt "20" as argument when calling its method on the instance
test.withdraw(101)

#works
```
